**Plotting the frequency of the country**

```python
def plot(givendata):
    cat =df[givendata]
    catvalue = cat.value_counts()
    plt.figure(figsize=(16,9))
    plt.bar(catvalue.index, catvalue,color=['red','blue','black','brown','yellow','pink','#008080'])
    plt.xticks(catvalue.index, catvalue.index.values)
    plt.ylabel("values")
    plt.title(givendata)
    
    plt.show()
    print("{}: \n {}".format(givendata,catvalue))
```

```python
df_cat=[ 'Country']

for i in df_cat:
    plot(i)
```

# Plotting the frequency of numerical category

```python
def plot(givendata):
    num =df[givendata]
    numvalue = num.value_counts()
    plt.figure(figsize=(16,9))
    plt.bar(numvalue.index, numvalue,color=['red','blue','black','brown','yellow','pink','#008080'])
    plt.xticks(numvalue.index, numvalue.index.values)
    plt.ylabel("values")
    plt.title(givendata)
    
    plt.show()
    print("{} : \n {}".format(givendata,numvalue))
```

```python
df_num1=['Career Span','Matches','Innings','Balls','Runs','Wickets','Average','Economy','Strike Rate','4','5','Best Figures']
for i in df_num1:
    plot(i)
```

# Analysis of wickets and balls of players

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plt.title("WICKETS AND RUNS OF PLAYERS")
plt.plot(df['Runs'],label='Runs',color='red', marker='*')
plt.plot(df['Balls'],label='Balls',color='b',marker='.')
plt.xlabel("Player",fontsize="16")
plt.legend(loc=2,fontsize=20)
```

# Analysis of matches and innings of players

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plt.title("MATCHES AND INNINGS OF PLAYERS")
plt.plot(df['Matches'],label='MATCHES',color='c', marker='*')
plt.plot(df['Innings'],label='INNINGS',color='m',marker='.')
plt.xlabel("Player",fontsize="16")
plt.legend(loc=2,fontsize=20)
```

# Analysis of average and economy of players

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plt.title("AVERAGE AND ECONOMY OF PLAYERS")
plt.plot(df['Average'],label='Averages',color='y', marker='*')
plt.plot(df['Economy'],label='Economy',color='b',marker='.')
plt.xlabel("Player",fontsize="16")
plt.legend(loc=2,fontsize=20)
```
